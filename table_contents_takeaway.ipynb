```python
# import the libraries
import sqlite3
import pandas as pd

# create a connection
conn = sqlite3.connect("databases/takeaway.db")
```

Tabla Restaurants

```python
query1 = """SELECT primarySlug, restaurant_id, name, city
           FROM restaurants;
"""

df_1  = pd.read_sql_query (query1, conn)
print(df_1)
```
